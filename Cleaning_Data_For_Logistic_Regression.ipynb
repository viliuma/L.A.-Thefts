```python
from google.colab import drive
import pandas as pd
import numpy as np
#import formatting tools
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Data_For_Collab/Crime_Data_from_2020_to_Present.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df = df[(df['date_reported'] >= '2022-01-01') & (df['date_reported'] < '2023-01-01')].copy()
```

```python
df.columns
```

```python
columns_to_drop = ['division_number', 'date_occurred','reporting_district', 'part', 'crime_code', 'premise_code','weapon_code',
       'weapon_description', 'status', 'crime_code_1',
       'crime_code_2', 'crime_code_3', 'crime_code_4', 'cross_street', 'latitude', 'longitude', ]
df = df.drop(columns=columns_to_drop)
```

```python
df['crime_description'].value_counts()
```

```python
pd.set_option('display.max_rows', None)
df['premise_description'].value_counts()
```

```python
df['THEFT OF IDENTITY'] = df['crime_description'].apply(lambda x: 1 if x == 'THEFT OF IDENTITY' else 0)
df['SINGLE FAMILY DWELLING'] = df['premise_description'].apply(lambda x: 1 if x == 'SINGLE FAMILY DWELLING' else 0)
df['MULTI-UNIT DWELLING (APARTMENT, DUPLEX, ETC)'] = df['premise_description'].apply(lambda x: 1 if x =='MULTI-UNIT DWELLING (APARTMENT, DUPLEX, ETC)' else 0)
df['BANK'] = df['premise_description'].apply(lambda x: 1 if x == 'BANK' else 0)
df['AUTOMATED TELLER MACHINE (ATM)'] = df['premise_description'].apply(lambda x: 1 if x == 'AUTOMATED TELLER MACHINE (ATM)' else 0)
df['CYBERSPACE'] = df['premise_description'].apply(lambda x: 1 if x == 'CYBERSPACE' else 0)
df['OTHER BUSINESS'] = df['premise_description'].apply(lambda x: 1 if x ==  'OTHER BUSINESS' else 0)
df['OTHER PREMISE'] = df['premise_description'].apply(lambda x: 1 if x == 'OTHER PREMISE' else 0)
```

```python
df.head()
```

```python
columns_to_drop = ['date_reported',
                  'area',
                   'area_name',
                   'modus_operandi',
                   'victim_age',
                   'victim_sex',
                   'victim_descent',
                   'status_description',
                   'location']

# Drop the specified columns
df = df.drop(columns=columns_to_drop).reset_index(drop=True)
```
